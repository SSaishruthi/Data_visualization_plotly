# Plotly 

Plotly is the interactive visualization library which is connected to pandas using cufflinks

```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#To connect plotly to notebook
init_notebook_mode(connected=True)
```

```python
#To use cufflinks offline
cf.go_offline()
```

```python
df1 =pd.DataFrame(np.random.randn(200,5), columns='A B C D E'.split())
```

```python
df1.head()
```

```python
df2 =  pd.DataFrame({'Book':['A','B','C'], 'Range':[25,67,95]})
```

```python
df2
```

```python
df1.plot()
```

```python
#To convert this simple plot into plotly interactive plot
df1.iplot()
```

# scatter plot

```python
df1.iplot(kind='scatter', x='A', y='B', mode='markers')
```

# Barplot

```python
df2.iplot(kind='bar', x='Book', y='Range')
```

```python
#Bar plot representing the sum of each column in df1
df1.sum().iplot(kind='bar')
```
